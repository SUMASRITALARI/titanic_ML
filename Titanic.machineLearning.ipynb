```python
df['Age'] = df['Age'].fillna(df['Age'].median())          
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])  

df = df.drop(columns=['Cabin'], errors='ignore')
df.isnull().sum()
```

```python

df['Sex'] = df['Sex'].astype(str).str.strip().str.lower()
df['Sex']=df['Sex'].map({'male':0,'female':1})
print(df['Sex'])
```

```python
df = pd.get_dummies(df, columns=['Embarked'])
```

```python
print(df.columns) 
```

```python
features = ['PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Embarked_C', 'Embarked_Q', 'Embarked_S' ]

X= df[features]
y= df['Survived']

            
```

```python
X=df.drop(['Name','Ticket','Survived'],axis=1)
y=df['Survived']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

model = RandomForestClassifier(random_state=42)
scores = cross_val_score(model, X, y, cv=5)  # 5-fold CV

print("Mean Accuracy:", scores.mean())
print("All fold scores:", scores)
```

```python
test = pd.read_csv(r"C:\Users\T SUMASRI\Downloads\titanic(1)\test.csv")
```

```python
missing_cols = set(df.drop(['Survived'], axis=1).columns) - set(test.columns)
for c in missing_cols:
    test[c] = 0

test = test[df.drop(['Survived'], axis=1).columns]
```

```python
from sklearn.ensemble import RandomForestClassifier
X = df.drop(['Survived', 'Name', 'Ticket'], axis=1)
y = df['Survived']
model = RandomForestClassifier(random_state=42)
model.fit(X, y)
```

```python

test['Sex'] = test['Sex'].astype(str).str.strip().str.lower()
test['Sex'] = test['Sex'].map({'male': 0, 'female': 1})
```

```python
print(test['Sex'].unique())
```

```python
test = test.drop(['Name', 'Ticket'], axis=1, errors='ignore')


predictions = model.predict(test)

```

```python
submission = pd.DataFrame({
    'PassengerId': test['PassengerId'],
    'Survived': predictions
})

submission.to_csv('submission.csv', index=False)
print(" submission.csv file created successfully!")
```
